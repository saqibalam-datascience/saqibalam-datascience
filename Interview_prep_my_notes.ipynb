```python
# Pascal Triangle
from math import factorial

n=int(input())

for x in range(n):
  for y in range(n-x):
    print(end= ' ')
  for y in range(x+1):
    print(factorial(x)//(factorial(x-y)*factorial(y)), end = ' ')
  print()
```

```python
# Armstrong number
# 1^3 + 5^3 + 3^3 = 153
num = int(input())
order = len(str(num))
sum = 0
temp = num
while temp>0:
  digit = temp%10
  sum += digit**order
  temp = temp//10
if sum == num:
  print('Yes')
else:
  print('No')
```

```python
# Perfect Number
num = int(input())
sum=0
temp = num
half = temp//2
for i in range(1,half+1):
  if (num%i == 0):
    sum +=i
if sum == num:
  print('Yes')
else:
  print('No')
```

```python
# OUTPUT
'''
cust_id | interpurchase(in days)
'''
import pandas as pd
data = pd.DataFrame({
    'date' : ['1/1/22','1/2/22', '3/2/22', '10/2/22', '5/3/22', '6/3/22'],
    'cust_id' : ['a','b','c','a','a','b'],
    'amount' : [100,200,332,122,233,100]
})
data.date = pd.to_datetime(data.date)
data
```

```python
def recur_fibo(n):
   if n <= 1:
       return n
   else:
       return(recur_fibo(n-1) + recur_fibo(n-2))

nterms = 10

# check if the number of terms is valid
if nterms <= 0:
   print("Plese enter a positive integer")
else:
   print("Fibonacci sequence:")
   for i in range(nterms):
       print(recur_fibo(i))
```

```python
# Write a function to flatten a nested list (e.g. [1, [2, [3, 4]]]).
l=[1, [2, [3, 4]]]
def flatten_list(li):
    flat = []
    for i in li:
        if isinstance(i, list):
            flat.extend(flatten_list(i))
        else:
            flat.append(i)
    return flat

flatten_list(l)
```
